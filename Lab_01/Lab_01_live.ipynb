```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stat 
```

### Loading data

```python
# Load temperature measurement data as an array
temp_measured = np.genfromtxt('Lab01 Temperature Data.csv', delimiter = ',')

# Load efficiency measurement data as an array
eff_measured = np.genfromtxt('Lab01 Solar Panel Efficiency Data.csv', delimiter = ',') 
```

### Plot histograms

```python
figure1 = plt.figure()
plt.hist(temp_measured, bins = 10)
plt.xlabel("Temperature")
plt.ylabel('Frequency')
plt.grid()
```

```python
figure2 = plt.figure()
plt.hist(eff_measured, bins = 10)
plt.xlabel("Efficiency")
plt.ylabel('Frequency')
plt.grid()
```

## 2) Compare the mean and median values for each of the temperature and efficiency 
datasets

```python
# Temperature statistics
temp_mean = np.mean(temp_measured)
temp_median = np.median(temp_measured)
temp_std = np.std(temp_measured)
temp_iqr = stat.iqr(temp_measured)
print('Mean temperature: ', temp_mean)
print('Median temperature: ', temp_median)
print('Standard deviation temperature: ', temp_std)
print('IQR temperature: ', temp_iqr)
# Efficiency statistics
eff_mean = np.mean(eff_measured)
eff_median = np.median(eff_measured)
eff_std = np.std(eff_measured)
eff_iqr = stat.iqr(eff_measured)
print('Mean efficiency: ', eff_mean)
print('Median efficiency: ', eff_median)
print('Standard deviation efficiency: ', eff_std)
print('IQR efficiency: ', eff_iqr)
```

### 5) Generate theoretical histograms that simulate the panel operating specifications.### 
Compare your result with the histograms for the measured datasets.


```python
mu_T, sigma_T, N_T = 15, 5, 1095
temp_simulated = np.random.normal(mu_T, sigma_T, N_T)
```

```python
fig = plt.figure()
plt.hist(temp_measured, bins = 10, histtype='bar', ec='black', label='Measured')
plt.hist(temp_simulated, bins = 10, histtype='bar', ec='black', label='Simulated')
plt.ylabel('Frequency')
plt.xlabel('Temperature')
plt.legend()
```

```python
mu_eff, sigma_eff, N_eff = 17, 2, 1095
eff_simulated = np.random.normal(mu_eff, sigma_eff, N_eff)

fig = plt.figure()
plt.hist(eff_measured, bins = 10, histtype='bar', ec='black', label='Measured')
plt.hist(eff_simulated, bins = 10, histtype='bar', ec='black', label='Simulated')
plt.axvline(x = np.mean(eff_measured), ls = 'dashed', color = 'r')
plt.axvline(x = np.mean(eff_simulated), ls = 'dashed', color = 'limegreen')
plt.ylabel('Frequency')
plt.xlabel('Efficiency')
plt.legend()
```
