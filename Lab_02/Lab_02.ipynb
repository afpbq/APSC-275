#### Load Python libraries

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stat
```

#### Load data

```python
# Load temperature measurement data as an array
tree_measured = np.genfromtxt('Lab02 Tree Population Data.csv', delimiter = ',')
```

#### Simulate tree dataset

```python
mu = 493
tree_simulated = np.random.poisson(mu, 100)
tree_simulated
```

```python
# Figure comparing measured and simulated tree
# Compare simulated data with reference data with a bar graph of PDF
# (Part 2)
# Construct a single array of the old and new data for looping
tree_data = np.vstack([tree_measured, tree_simulated]) 
datalabels = ['Old','New'] # Define legend labels for plot
expected_value = [] # Define varaible for holding expected value calculations
standard_deviation = [] # Define variable for holding standard deviaiton calculations
N = len(tree_data[0])
print('Part 2')
fig1 = plt.figure()
ax0 = plt.subplot(111)
# plt.hold(True)
for i in range(len(tree_data)): # Loops over each dataset
    #Extract frequency data from datasets
    value, frequency = np.unique(tree_data[i], return_counts=True) 
    
    # Calculate expected value for datasets (part 3)
    expected_value.append(np.dot(value, frequency) / N)

    # Calculate standard deviation for datasets (part 3)
    standard_deviation.append(np.sqrt(np.sum(np.square(value) * frequency / N) - np.square(expected_value[i])))

    # Plot bar graph
    plt.bar(value, frequency/N, ec='black', label=datalabels[i])

# Customize plot
plt.title(r"$\mathrm{PDF\/\/of\/\/Old\/\/&\/\/New\/\/Data}$") #Add title
plt.xlabel(r"$\mathrm{Healthy\/\/Trees}$") #Add horizontal axis label
plt.ylabel(r"$\mathit{P}(\mathrm{Healthy\/\/Trees})$") #Add vertical axis label
plt.tick_params(direction='in') #Move tick marks inside axes
plt.legend()
plt.show() #Display bar plot

print('Part 3')
print(expected_value)
print(standard_deviation)
```

```python
# Compare simulated data with reference data with a bar graph of CDF
# (Part 4)
print('Part 4')
median = [] # Define variable for holding median calculations
IQR = [] # Define variable for holding IQR calculations
Over_500=[] # Define variable for holding calculation of forest sections with over 500 trees on them
for i in range(len(tree_data)): # Loops over each data set
    #Extract frequency data from datasets
    value, frequency = np.unique(tree_data[i], return_counts=True) 

    # Calculate the CDF for number of trees in each forest section
    cumulative_frequency = np.cumsum(frequency/N) 

    # Calculate median for datasets (part 5)
    median.append(np.median(tree_data[i])) 

    # Calculate IQR for datasets (part 5)
    IQR.append(stat.iqr(tree_data[i])) 

    # Calculate number of forest sections with over 500 trees (part 6)
    Over_500.append((tree_data[i] > 500).sum()) 

    # Plot cumulative frequency of each dataset
    plt.bar(value, cumulative_frequency, ec='black', label=datalabels[i])

#Plot bar graph
plt.title(r"$\mathrm{CDF\/\/of\/\/Old\/\/&\/\/New\/\/Data}$") #Add title
plt.xlabel(r"$\mathrm{Healthy\/\/Trees}$") #Add horizontal axis label
plt.ylabel(r"$\mathit{F}(\mathrm{Healthy\/\/Trees})$") #Add vertical axis label
plt.legend()
plt.show() #Display bar plot


print('Part 5')
print(median)
print(IQR)
print('Part 6')
print(Over_500)
```
