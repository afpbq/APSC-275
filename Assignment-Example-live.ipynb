```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stat
import pandas as pd
```

```python
# Load data into a pandas dataframe
df = pd.read_excel('trees-municipal.xlsx', header = 0)
print(df)
```

```python
# Cleaning dataset
# Assign NaN to 999 values
df[df['Trunk Diameter (cm)']==999] = np.NaN

# Remove NaN data
df = df.dropna()

# Remove non-usable information
df = df.drop(columns=['Owner', 'Scientific Name'])
print(df)
```

```python
# Group data by park
df_grouped = df.groupby(['Park'])
print(df_grouped)
# len(df_grouped)
groups = dict(list(df_grouped))
groups

# Count elements in each group
group_counts = df_grouped.count()
group_counts.sort_values(by = 'Common Name', ascending = False, inplace = True)
group_counts
```
